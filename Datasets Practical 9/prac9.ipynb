```python
# Import clusters file
import clusters
import PIL


# Read data and store it in memory
[rownames, colnames, data] = clusters.readfile('blogdata.txt')

#print('stored data is ',data)



clust = clusters.hcluster(data)

# Print the cluster
#clusters.printclust(clust)

# draw dendrogram from cluster
print("aaaaaaaaaaaaaaa")
dendrogramcluster = clusters.drawdendrogram(clust, rownames,"clust.jpg")

#colnames
#clusters.printclust(dendrogramcluster)

kcl=clusters.kcluster(rows=data,k=10)

scd=clusters.scaledown(data)
clusters.draw2d(scd,rownames)
#string(kcl)

rotm=clusters.rotatematrix(data)
rotclust=clusters.hcluster(rotm)
rotdendrogramcluster = clusters.drawdendrogram(rotclust, colnames,"rotclust.jpg")
```
