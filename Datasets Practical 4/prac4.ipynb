```r
library("ggplot2") 
library("plyr")      # for replace
library("rgdal")     # for readOGR and other vector maps and spatial objects
library("maptools")  # for fo

mydata<- read.csv("whodata.csv")
names(mydata)<-c("WHO", names(mydata)[2:11])
myworld<-readOGR(dsn = ".", layer = "ne_110m_admin_0_countries")
myworld.f<- fortify(myworld, region = "sov_a3")
myworld.f <-merge(myworld.f, myworld@data, by.x="id", by.y="sov_a3")
gmyworld <-ggplot(myworld.f, aes(long, lat, group=group, fill=continent))+geom_polygon()+coord_equal()
#gmyworld
myworld.f<-merge(myworld.f, mydata, by.x="sovereignt",  by.y = "WHO")
myworld.f.1990<-myworld.f[myworld.f$X.1 == 1990, ]
myworld.f.2013<-myworld.f[myworld.f$X.1 == 2013, ]

gmyworld1990<-ggplot(myworld.f.1990, aes(long, lat, group=group, fill=as.numeric(as.character(myworld.f.1990[,71]))))+geom_polygon()+coord_equal()+scale_fill_gradient2(low = "red",high = "blue", mid = "yellow",midpoint=60,limits=c(30,90))
gmyworld2013<-ggplot(myworld.f.2013, aes(long, lat, group=group, fill=as.numeric(as.character(myworld.f.2013[,71]))))+geom_polygon()+coord_equal()+scale_fill_gradient2(low = "red",high = "blue", mid = "yellow",midpoint=60,limits=c(30,90))
gmyworld1990
gmyworld2013
#names(myworld.f)
```
