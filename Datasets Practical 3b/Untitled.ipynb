```r
outcome <- read.csv("outcome-of-care-measures.csv") 
library(scales)
library(reshape2)
library("ggplot2")

outcome[,11]=as.numeric(outcome[,11])

plot1<-ggplot(outcome, aes(outcome[,11], ..density..) )
plot1<-plot1+geom_histogram(binwidth = 5)
plot1<-plot1+geom_density(color = "red")
plot1<-plot1+ylab("Densidad")+xlab("Ratio de muertes en 30 días")+ggtitle("Ataques al corazón. Ratio de muertes en 30 días")+theme(plot.title = element_text(hjust = 0.5))
plot1<-plot1+scale_x_continuous(limits=c(1,95))
plot1
```

```r
outcome <- read.csv("outcome-of-care-measures.csv") 
library(scales)
library(reshape2)
library("ggplot2")
outcomesel<-outcome[,c(11,17,23)]
colnames(outcomesel)[1:3]<-c("Heart_Attack","Heart_Failure","Pneumonia")

outcomesel$Heart_Attack=as.numeric(outcomesel$Heart_Attack)
outcomesel$Heart_Failure=as.numeric(outcomesel$Heart_Failure)
outcomesel$Pneumonia=as.numeric(outcomesel$Pneumonia)

plot2<-ggplot(outcomesel, aes(outcomesel$Heart_Attack) )
plot2<-plot2+geom_density(aes(fill="Heart_Attack"),adjust=1, alpha=1/5)
plot2<-plot2+geom_density(data=outcomesel, aes(outcomesel$Heart_Failure,fill="Heart_Failure"), alpha=1/5)
plot2<-plot2+geom_density(data=outcomesel, aes(outcomesel$Pneumonia,fill="Pneumonia"), alpha=1/5)
plot2<-plot2+scale_x_continuous(limits=c(0,80))+xlab("Ratio de muertes en 30 días")+scale_fill_hue("Tipología")

#plot2<-plot2+theme(legend.position = c(0, 1),legend.justification = c(0, 1))

plot2
```

```r
outcome <- read.csv("outcome-of-care-measures.csv") 
library(scales)
library(reshape2)
library("ggplot2")

t<-table(outcome$State)
outcome2<-subset(outcome, t[outcome$State]>20)
outcome3<-subset(outcome2, outcome2[,11]!="Not Available")
outcome3[,11]=as.numeric(outcome3[,11])
plot3<-ggplot(outcome3, aes(x=outcome3$State, y=outcome3[,11]))
plot3<-plot3+geom_boxplot()
plot3<-plot3+theme(axis.text.x=element_text(angle=90))+ylab("Ratio de muertes en 30 días")+xlab("Estado")+ggtitle("Ratio de muertes por ataque al corazón")+theme(plot.title = element_text(hjust = 0.5))
plot3
```
